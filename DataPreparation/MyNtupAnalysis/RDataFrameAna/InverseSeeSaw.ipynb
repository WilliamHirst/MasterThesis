```python
import ROOT as R
import array
%jsroot on
```

```python
list = []
for vln in range(1,101):
    list.append(vln/10000.)
vln = array.array('f',list)

list = []
for mn in range(1,1050,50):
    list.append(mn)
mn = array.array('f',list)
```

```python
vln
```

```python
h2 = R.TH2F("nm_to_wl","nm_to_wl",len(vln),vln[0],vln[-1],len(mn),mn[0],mn[-1])
wmass = 80.379
zmass = 91.187
hmass = 125.35
ar_nm_to_wl = array.array('f')
ar_nm_to_zl = array.array('f')
ar_nm_to_hl = array.array('f')
for i in range(len(vln)):
    for j in range(len(mn)):
        rw = (wmass/mn[j])**2
        rz = (zmass/mn[j])**2
        rh = (hmass/mn[j])**2
        nm_to_wl = 2*(vln[i]**2)*(mn[j]**3)*((1-rw)**2)*((1+2*rw))
        nm_to_zl = (vln[i]**2)*(mn[j]**3)*((1-rz)**2)*((1+2*rz))
        nm_to_hl = (vln[i]**2)*(mn[j]**3)*((1-rh)**2)
        total = nm_to_wl+nm_to_zl+nm_to_hl
        if total > 0:
            h2.Fill(vln[i],mn[j],nm_to_wl/total)
        #print(total)
        #if total:
        if i == 3:
            print(vln[i],nm_to_wl/total)
            ar_nm_to_wl.append(nm_to_wl/total)
            ar_nm_to_zl.append(nm_to_zl/total)
            ar_nm_to_hl.append(nm_to_hl/total)
        
```

```python
c = R.TCanvas()
c.Draw()
h2.Draw("colz")
```

```python
gr_nm_to_wl = R.TGraph(len(mn),mn,ar_nm_to_wl)
gr_nm_to_zl = R.TGraph(len(mn),mn,ar_nm_to_zl)
gr_nm_to_zl.SetLineColor(R.kAzure+7)
gr_nm_to_hl = R.TGraph(len(mn),mn,ar_nm_to_hl)
gr_nm_to_hl.SetLineColor(R.kSpring+7)
```

```python
c1 = R.TCanvas("c1","c1",1)
c1.Draw()
gr_nm_to_hl.Draw("AC*")
gr_nm_to_hl.SetMaximum(1.0)
gr_nm_to_hl.SetMinimum(1.0)
gr_nm_to_wl.Draw("C*")
gr_nm_to_zl.Draw("C*")
```
